```python
import pandas as pd
import ipywidgets as widgets
from pandas import ExcelWriter
import os
```

```python
df=pd.read_excel("timesheet_sample file.xls",sheet_name="All projects",skiprows=7)
```

```python
df=df.fillna(0)
```

```python
df=df.iloc[0:10160,:]
```

```python
df
```

```python
projects=df['Project Name'].unique()
```

```python
with pd.ExcelWriter('Projects1.xlsx') as writer:  
    
    for i in range(len(projects)):
        p= projects[i]
        newdf=df[df['Project Name']==p]
        p=p.replace(":","")
        sheetname=p[:30]
        newdf.to_excel(writer,sheet_name=sheetname)
        newdf.style.apply(lambda x: ["background-color: #ff33aa" 
                          if (i >= 6 and (v =='Billable')) 
                          else "" for i, v in enumerate(x)], axis = 1)
        del sheetname
        del newdf
```

```python
sm=df[df['Project Name']=='YogaBar - Amazon Marketing']
```

```python
sm['Date']=pd.to_datetime(sm['Date'],format='%d-%m-%Y')
```

```python
sm.info()
```

```python
sm['Month'] = sm['Date'].map(lambda x:  x.month_name)
```

```python
sm['Hours(For Calculation)'] = sm['Hours(For Calculation)'].map(lambda x:  float(x))
```
